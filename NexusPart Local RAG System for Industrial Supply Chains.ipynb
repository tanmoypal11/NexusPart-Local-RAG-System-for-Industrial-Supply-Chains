# 3. Cleaning & Normalization

```python
# 1. Fill Missing Values
# For technical numbers, 0 is safer. For text, 'N/A' is better.
df['DESCRIPTION'] = df['DESCRIPTION'].fillna('No description available')
df = df.fillna('0') 

# 2. Text Normalization
# Convert everything to lowercase to avoid "Bolt" vs "bolt" search errors
text_cols = ['DESCRIPTION', 'Attribut1', 'Application', 'Characteristic']
for col in text_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()

# 3. Standardization (Regex)
# Ensure "5 X 20mm" and "5x20 mm" become "5x20mm"
df['DESCRIPTION'] = df['DESCRIPTION'].str.replace(r'\s*x\s*', 'x', regex=True)
```

# 4. Create the "Context String"

```python
def create_context(row):
    return (f"Part ID {row['ID']} is a {row['Characteristic']} {row['Material']} component. "
            f"Description: {row['DESCRIPTION']}. "
            f"Rated Current: {row['Rated Current (A)']}A, Rated Voltage: {row['Rated Voltage (V)']}V. "
            f"Application: {row['Application']}.")

df['combined_text'] = df.apply(create_context, axis=1)
```

# 5. Generate Embeddings

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. Load the lightweight "Brain"
# This will download about 80MB the first time you run it
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Convert your text column into a list of strings
# We must ensure there are no NaNs in this column before encoding
sentences = df['combined_text'].astype(str).tolist()

# 3. Encode the text into Vectors
# This is the "Heavy Lifting" part. 
# On 998 rows, this should take about 5-10 seconds on a CPU.
embeddings = model.encode(sentences, show_progress_bar=True)

# 4. Verify the shape
print(f"Number of parts encoded: {embeddings.shape[0]}")
print(f"Dimensions per part: {embeddings.shape[1]}") # Should be 384
```

## Sanity Check

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compare the first part with the second part
sim = cosine_similarity([embeddings[0]], [embeddings[1]])
print(f"Similarity between first two parts: {sim[0][0]:.2f}")
```

```python
import chromadb
import os

# 1. Define your professional project path
# We use 'r' to ensure Windows backslashes don't cause errors
project_path = r"E:\Desktop\GUVI\Project\NexusPart Local RAG System for Industrial Supply Chains"
db_path = os.path.join(project_path, "nexus_part_db")

# 2. Initialize the Persistent Client
# This creates the physical database folder on your E: drive
client = chromadb.PersistentClient(path=db_path)

# 3. Create or Get the Collection
# We'll use 'cosine' similarity as we verified in your last step
collection = client.get_or_create_collection(
    name="industrial_components",
    metadata={"hnsw:space": "cosine"}
)

# 4. Add your data
# We include the vectors, the original text, and metadata for filtering
# (Adding a few key metadata fields like Material and Rating)
collection.add(
    ids=df['ID'].astype(str).tolist(),
    embeddings=embeddings.tolist(),
    documents=df['combined_text'].tolist(),
    metadatas=[{
        "material": str(m), 
        "rating": str(r),
        "application": str(a)
    } for m, r, a in zip(df['Material'], df['Rating'], df['Application'])]
)

print(f"✅ Success! Database saved at: {db_path}")
print(f"Total parts stored: {collection.count()}")
```

# 8. Build the RAG Pipeline

```python
import ollama
import chromadb

# --- CONNECT TO CHROMA ---
client = chromadb.Client(
    chromadb.config.Settings(
        persist_directory="E:/chroma_parts_db"
    )
)

collection = client.get_or_create_collection(name="industrial_parts")


def get_nexus_recommendation(user_query):
    results = collection.query(
        query_texts=[user_query],
        n_results=3
    )
    
    retrieved_docs = results['documents'][0]
    context = "\n\n".join(retrieved_docs)
    
    system_prompt = (
        "You are a Senior Industrial Procurement Engineer. "
        "Use the following technical data to answer the user's request. "
        "If a part is a good substitute, explain WHY based on specs. "
        "If unsure, warn user.\n\n"
        "TECHNICAL DATA:\n" + context
    )
    
    response = ollama.chat(
        model='llama3:8b',
        messages=[
            {'role': 'system', 'content': system_prompt},
            {'role': 'user', 'content': user_query}
        ],
        options={'num_ctx': 2048, 'temperature': 0.2}
    )
    
    return response['message']['content'], retrieved_docs


# --- TEST ---
query = "I need a substitute for a 1.6A ceramic fuse used in primary protection."
answer, parts_found = get_nexus_recommendation(query)

print(answer)
```

## app.py

```python
import streamlit as st
import chromadb
import ollama

# --- 1. CONFIGURATION ---
st.set_page_config(page_title="NexusPart Smart Search", layout="wide")
st.title("🛠️ NexusPart: Industrial Command Center")

# Your E: drive path
DB_PATH = r"E:\Desktop\GUVI\Project\NexusPart Local RAG System for Industrial Supply Chains\nexus_part_db"

# --- 2. DATA LOADERS ---
@st.cache_resource
def get_collection():
    client = chromadb.PersistentClient(path=DB_PATH)
    return client.get_collection(name="industrial_components")

collection = get_collection()

# Get all IDs for the dropdown list
def get_all_ids():
    data = collection.get()
    return sorted(data['ids'])

part_ids = get_all_ids()

# --- 3. THE RAG ENGINE ---
def run_nexus_engine(user_input):
    # Retrieve top 3 matches from ChromaDB
    results = collection.query(query_texts=[user_input], n_results=3)
    context = "\n\n".join(results['documents'][0])
    
    # Generate Engineering Response via Ollama
    response = ollama.chat(
        model='llama3:8b',
        messages=[
            {'role': 'system', 'content': f"You are an Industrial Engineer. Analyze this data: {context}"},
            {'role': 'user', 'content': user_input}
        ],
        options={'num_ctx': 2048, 'temperature': 0.1}
    )
    return response['message']['content'], results['documents'][0]

# --- 4. THE "SINGLE BOX" UI ---
# We use a selectbox that allows typing NEW text (accept_new_options=True)
st.markdown("### 🔍 Search or Select Part")
query = st.selectbox(
    "Type a Part ID or ask a technical question:",
    options=part_ids,
    index=None,
    placeholder="Example: 'A102' or 'Find ceramic substitutes for 1.6A fuses'...",
    accept_new_options=True  # THIS MERGES SEARCH AND DROPDOWN
)

col1, col2 = st.columns([2, 1])

with col1:
    if query:
        st.info(f"Processing: **{query}**")
        with st.spinner("Analyzing Database..."):
            answer, docs = run_nexus_engine(query)
            st.markdown("---")
            st.subheader("AI Analysis")
            st.write(answer)
            st.session_state.last_docs = docs
    else:
        st.write("Waiting for input... Use the box above to start.")

with col2:
    st.subheader("📋 Source Data")
    if "last_docs" in st.session_state:
        for i, doc in enumerate(st.session_state.last_docs):
            with st.expander(f"Reference Part {i+1}"):
                st.write(doc)
    else:
        st.info("Technical specs will appear here once you search.")
```
